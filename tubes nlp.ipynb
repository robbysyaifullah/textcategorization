```python
from twitterscraper import query_tweets


# All tweets matching either Trump or Clinton will be returned. You will get at
# least 10 results within the minimal possible time/number of requests
for tweet in query_tweets("Pemerintahan OR Presiden", 10)[:10]:
    print(tweet.text.encode('utf-8'))
```

```python
list_of_tweets = query_tweets("Presiden OR Jokowi OR Pemerintah OR Menteri", 100)

#print the retrieved tweets to the screen:
for tweet in query_tweets("Presiden OR Jokowi OR Pemerintah OR Menteri", 100):
    print(tweet.text)

#Or save the retrieved tweets to file:
file = open('output.txt','w')
for tweet in query_tweets("Presiden OR Jokowi OR Pemerintah OR Menteri", 100):
    file.write(str(tweet.text.encode('utf-8')) +',\n')
file.close()
```

```python
import codecs, json
import pandas as pd

with codecs.open('tweets.json', 'r', 'utf-8') as f:
    tweets = json.load(f, encoding='utf-8')

list_tweets = [list(elem.values()) for elem in tweets]
list_columns = list(tweets[0].keys())
df = pd.DataFrame(list_tweets, columns=list_columns)
df
```
